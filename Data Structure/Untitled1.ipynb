```python
def sumOfMultipleOf(x,y):
    sum =0
    for i in range(1,1000):
        if(i%x==0 or i%y==0):
            sum += i
    print(f'The sum of multiples of {x} or {y} below 1000 is: {sum}')    

    
```

```python
sumOfMultipleOf(3,5)
```
